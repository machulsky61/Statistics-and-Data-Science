#(b) Bootstrap (intervalo de confianza de la mediana de WEIG)

```r
#Bootstrap Normal
estimar_se_mediana <- function(data, N = 1000){
  titahatboot <- rep(0, N)
  for(i in 1:N){
    Xboot <- sample(data, length(data), replace = TRUE)
    titahatboot[i] <- median(Xboot)
  }
  sqrt(mean((titahatboot-mean(titahatboot))^2))
}

#Intervalo bootsrap normal
bootstrap_normal <- function (data, n) {
  mediana_muestra <- median(data)
  se_boot <- estimar_se_mediana(data, n)
  intervalo_boot <- c(mediana_muestra - 1.96*se_boot, mediana_muestra + 1.96*se_boot)
  return(intervalo_boot)
}



#Intervalo Bootstrap Percentil
bootstrap_percentil <- function(data, n) {
  mediana_original <- median(data)
  bootstrap_samples <- matrix(sample(data, n * length(data), replace = TRUE), n)
  mediana_bootstrap <- apply(bootstrap_samples, 1, median)

  sesgo_bootstrap <- mediana_bootstrap - mediana_original  # Corrección del sesgo
  intervalo_confianza <- quantile(sesgo_bootstrap, c(0.025, 0.975))

  return(mediana_original + intervalo_confianza)
}
```

#(e) Leave one out para hallar ventana óptima

```r
loocv_bw <- function(data, x_col, y_col, bw_values) {
  n <- nrow(data)
  mse_values <- numeric(length(bw_values))

  for (i in seq_along(bw_values)) {
    mse <- 0

    for (j in 1:n) {
      # Deja fuera un punto en cada iteración para LOOCV
      data_leave_one_out <- data[-j, ]

      # Ajusta el modelo con ksmooth
      smooth_fit <- ksmooth(data_leave_one_out[[x_col]], data_leave_one_out[[y_col]],kernel = "normal", bandwidth = bw_values[i], x.points = data[j, ][[x_col]])

      # Predice el valor para el punto excluido
      predicted_value <- smooth_fit$y[1]

      y_facto = data[j, ][[y_col]]


    #  if(is.na(y_fact)){print('yfact es NA')} #debug
    #  if(is.na(predicted_value)){print(cat('predicted_value es NA','y j es:',j))}

      # Calcula el error cuadrático
      #mse <- mse + (y_fact - predicted_value)^2

      if(!is.na(y_facto) && !is.na(predicted_value)){
        # Calcula el error cuadrático
        mse <- mse + (y_facto - predicted_value)^2
      }
    }

    # Calcula el error cuadrático medio para el bandwidth
    #mse_values[i] <- mse / n
    mse_values[i] <- mse / sum(!is.na(data[, y_col]))
    #print(sum(!is.na(data[, y_col])))
  }

  # Encuentra el índice del bandwidth con el menor error cuadrático medio
  best_bw_index <- which.min(mse_values)
  best_bw <- bw_values[best_bw_index]

  return(list(best_bandwidth = best_bw, mse_values = mse_values))
}
```

# (h) Nuevo modelo con algunas variables seleccionadas

```r
modelizar <- function (predictores, respuesta, datos_entrenamiento, datos_prueba) {
    # Crea el modelo lineal
    modelo <- lm(formula = paste(respuesta, "~", paste(predictores, collapse = "+")), data = datos_entrenamiento)

    # Muestra un resumen del modelo
    print(summary(modelo))

    #A partir de acá se usan datos_prueba:

    # Realiza predicciones en los datos de prueba
    predicciones_prueba <- predict(modelo, newdata = datos_prueba)

    # Calcular el error de predicción empírico (error cuadrático medio)
    error_empirico <- mean((datos_prueba$WEIG - predicciones_prueba)^2)

    # Calcula el R cuadrado
    r_cuadrado <- 1 - sum((datos_prueba$WEIG - predicciones_prueba)^2) / sum((datos_prueba$WEIG - mean(datos_prueba$WEIG))^2)

    # Calcula el número de predictores en el modelo
    num_predictores <- length(predictores)

    # Calcula el número de observaciones en los datos de prueba
    num_obs_prueba <- nrow(datos_prueba)

    # Calcula el R cuadrado ajustado
    r_cuadrado_ajustado <- 1 - ((1 - r_cuadrado) * (num_obs_prueba - 1)) / (num_obs_prueba - num_predictores - 1)

    # Muestra los resultados
    resultados_prueba <- data.frame(Predicted_WEIG = predicciones_prueba, R_cuadrado = r_cuadrado, R_cuadrado_ajustado = r_cuadrado_ajustado)
    #print(resultados_prueba)

    # Mostrar los resultados
    cat("R^2 en el grupo de testeo:", r_cuadrado, '\n')
    cat("R^2 ajustado en el grupo de testeo", r_cuadrado_ajustado, '\n')
    cat("Error de predicción empírico en el grupo de testeo:", error_empirico)
}
```

```r
#Todas las variables
predictores <- c("BIAC","BIIL","BITRO","CHEST1","CHEST2","ELBOW","WRIST", "KNEE","ANKLE","SHOUL","CHESTG","WAISTG","NAVEL","HIP","GLUTE","BICEP", "FLOREA","KNEEG","CALF","ANKLEG","WRISTG","AGE","HEIG","GEN")
respuesta <- 'WEIG'


# Ajusta un modelo LASSO con glmnet
modelo_lasso <- glmnet(as.matrix(datos_entrenamiento[, predictores]), datos_entrenamiento$WEIG)

# Imprime la tabla de coeficientes para distintos valores de lambda
print(coef(modelo_lasso))

# Grafica los coeficientes en función de lambda
plot(modelo_lasso, xvar = "lambda", label = TRUE)
```

# (j) Lambda óptimo

```r
tabla_coeficientes <- coef(modelo_lasso)
#print(tabla_coeficientes)

# Calcula y almacena el R cuadrado para cada valor de lambda en los datos de prueba
r_cuadrado_por_lambda <- numeric(length(modelo_lasso$lambda))
for (i in seq_along(modelo_lasso$lambda)) {
  predicciones_prueba <- predict(modelo_lasso, newx = as.matrix(datos_prueba[, predictores]), s = modelo_lasso$lambda[i])
  r_cuadrado_por_lambda[i] <- 1 - sum((datos_prueba$WEIG - predicciones_prueba)^2) / sum((datos_prueba$WEIG - mean(datos_prueba$WEIG))^2)
}


resultados_lasso <- data.frame(lambda = modelo_lasso$lambda, R_cuadrado = r_cuadrado_por_lambda)

par(mfrow = c(1, 2))
plot(modelo_lasso, xvar = "lambda")
plot(resultados_lasso$lambda, resultados_lasso$R_cuadrado, type = "l", xlab = "Lambda", ylab = "R cuadrado", main = "R cuadrado vs Lambda")
```

```r
cv_modelo <- cv.glmnet(as.matrix(datos_entrenamiento[, predictores]), datos_entrenamiento$WEIG, alpha = 1)

# Muestra el valor óptimo de lambda según el criterio de 1 desvío estándar
lambda_optimo <- cv_modelo$lambda.min
print(paste("Valor óptimo de lambda:", lambda_optimo))

# Ajusta el modelo con el valor óptimo de lambda
modelo_lasso_optimo <- glmnet(as.matrix(datos_entrenamiento[, predictores]), datos_entrenamiento$WEIG, alpha = 1, lambda = lambda_optimo)

# Imprime la tabla de coeficientes
print(coef(modelo_lasso_optimo))

# Calcula y almacena el R cuadrado en los datos de prueba
predicciones_prueba_optimo <- predict(modelo_lasso_optimo, newx = as.matrix(datos_prueba[, predictores]))
r_cuadrado_optimo <- 1 - sum((datos_prueba$WEIG - predicciones_prueba_optimo)^2) / sum((datos_prueba$WEIG - mean(datos_prueba$WEIG))^2)

# Calcula el error de clasificación empírico en el grupo de testeo
clases_predichas <- predict(modelo_lasso_optimo, newx = as.matrix(datos_prueba[, predictores]), s = lambda_optimo, type = "class")
error_clasificacion_optimo <- mean(clases_predichas != datos_prueba$WEIG)
print(paste("Error de clasificación empírico en el grupo de testeo:", error_clasificacion_optimo))
```
